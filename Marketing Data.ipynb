```python
# checking missing values 
print(df.isna().sum())
plt.figure(figsize = (15,6))
sns.heatmap(df.isnull())
plt.show()
```

 It is found that Income is the only column that have missing values

 To Clean the data 

The column ' Income ' has blank spaces on both sides which is to be removed 

```python
df.rename(columns = {' Income ':'Income'}, inplace = True)
```

Creating a function to remove the '$' sign from the Income and converting the dtype from 'object' to 'float'

```python
def converter(x):
    try: 
        return float(x.split('$')[1].replace(',',''))
    except AttributeError: # as there are some missing values 
        return np.NaN
    
df['Income'] =  df['Income'].apply(converter)   
```

Converting the dtype from 'object' to 'datetime64' so that we can create new columns for month and year of enrollment of the customers

```python
df['Dt_Customer'] = pd.to_datetime(df.Dt_Customer)
df['DT_month'] = df.Dt_Customer.dt.month
df['DT_year'] = df.Dt_Customer.dt.year
```

In the Education column
Basic - who passed school education
Graduation - who has completed a course for degree or diploma
2n Cycle - Not defined
Master - who obtain Master degree
PhD - who obtain PhD

To define 2n cycle - Merging with masters(As 2n cycle is somewhat equivalent to masters)

```python
print('Before:',df.Education.unique())
df.Education.replace({'2n Cycle':'Master'},inplace = True)
print('After:',df.Education.unique())
```

```python
df.Education.value_counts()
```

To further clean data 
Changing (Alone,YOLO, Absurd) into "Single" in Marital Status

```python
print(df.Marital_Status.unique())
df.Marital_Status.replace({'Alone':'Single','YOLO':'Single','Absurd':'Single'}, inplace = True)
print(df.Marital_Status.unique())
```
